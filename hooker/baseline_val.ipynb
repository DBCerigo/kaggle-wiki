```python
res_dict = {}

res_dict['visits_mean'] = np.round(np.mean(filled_train.drop('Page', axis=1).values, axis=1)).reshape(-1,1) # reshape to pass to smape as pred
res_dict['visits_med'] = np.round(np.median(filled_train.drop('Page', axis=1).values, axis=1)).reshape(-1,1)
res_dict['visits_med49'] = np.round(np.median(filled_train.drop('Page', axis=1).values[:, -49:], axis=1)).reshape(-1,1)
res_dict['visits_med30'] = visits_med30 = np.round(np.median(filled_train.drop('Page', axis=1).values[:,-30:], axis=1)).reshape(-1,1)
# Calculate median of last 49 days
res_dict['visits_nanmed49'] = np.nan_to_num(np.round(np.nanmedian(train.drop('Page', axis=1).values[:, -49:], axis=1)).reshape(-1,1))
```

```python
# nans in the validation set are apparently ignored: (https://www.kaggle.com/c/web-traffic-time-series-forecasting/discussion/36750)
def val_smape(true, pred):
    # array broadcasting should mean it's possible to (Nx1) dim pred.
    pred = np.nan_to_num(pred)
    diff = np.abs(true - pred).reshape(-1) # leave nans here
    nonna_inds = np.argwhere(~np.isnan(diff)).flatten()
    denom = (np.abs(true) + np.abs(pred)).reshape(-1) 
    smape = np.nan_to_num(diff[nonna_inds] / denom[nonna_inds]) # replace nans in case of div by 0
    return np.mean(smape) * 200
# doesn't ignore nans in the validation set
def fast_smape(true, pred, axis=None):
#     true = np.nan_to_num(true)
    raw_smape = np.abs(true - pred) / (np.abs(true) + np.abs(pred))
    kaggle_smape = np.nan_to_num(raw_smape)
    return np.mean(kaggle_smape, axis=axis) * 200
```

```python
true_vals = val.drop('Page', axis=1).values
true_vals.shape
res_dict['cheat_med'] = np.median(np.nan_to_num(true_vals), axis=1).reshape(-1,1)
```

```python
# val nans are ignored 47.74, 46.99
for baseline_name, preds in res_dict.items():
    print('{} :\t {:.2f}'.format(baseline_name, val_smape(true_vals, preds)))
```

```python
for baseline_name, preds in res_dict.items():
    print('{} :\t {:.2f}'.format(baseline_name, fast_smape(true_vals, preds)))
```

```python
val['median_pred'] = res_dict['visits_nanmed49']
val['median_smape'] = fast_smape(true_vals, res_dict['visits_nanmed49'], axis=1)
```
